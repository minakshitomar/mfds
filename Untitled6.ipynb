```python
import pandas as pd
import matplotlib.pyplot as plt
url='/content/IBM-HR-Analytics-Employee-Attrition-and-Performance-Revised (1).csv'
df= pd.read_csv(url)
df.dtypes
```

```python
df.head()
```

```python
df
```

```python
mean_value=df['TotalWorkingYears'].mean()
print(f"mean value:-{mean_value}")
```

```python
median_value=df['TotalWorkingYears'].median()
print(f"median value:-{median_value}")
```

```python
mode_value=df['TotalWorkingYears'].mode()
print(f"mode value:-{mode_value}")
```

```python
var_value=df['TotalWorkingYears'].var()
print(f"variance value:-{var_value}")
```

```python
sd_value=df['TotalWorkingYears'].std()
print(f"standard deviation value:-{sd_value}")
```

```python
skew_value=df['TotalWorkingYears'].skew()
print(f"skewness value:-{skew_value}")
```

```python
kur_value=df['TotalWorkingYears'].kurt()
print(f"kurtosis value:-{kur_value}")
```

```python
print(df.columns)
```

```python
average_performance=df.groupby('TotalWorkingYears')['TrainingTimesLastYear'].mean()
```

```python
average_performance.plot(kind='bar', color='skyblue')
#plt.title('Average Performance In Terms Of Years')
plt.xlabel('ToatalWorkingYears')
plt.ylabel('TraininTimesLast Year')
plt.show()
```
